```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

fig, axes = plt.subplots(2, 3, tight_layout=True)
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)
axes[0,0].set_title("折れ線グラフ")

x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)
axes[0,1].plot(x, y, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))
axes[0,1].set_title("サインカーブ")


x2 = np.random .normal(15, 5, 2000)
axes[0,2].hist(x2)
axes[0,2].set_title("ヒストグラム")

hertz = [300, 200, 200, 500, 600, 2000, 1600, 1800, 1850, 1700, 1500, 1500, 1500, 1400, 5000, 6000, 2000, 3000, 3000]
axes[1,0].boxplot(hertz)
axes[1,0].set_title('箱ひげ図')

np.random.seed(3)
x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))
axes[1,1].scatter(x4, y4)
axes[1,1].set_title("散布図")

x5 = np.array([30, 40, 30])
label = ["いぬ", "ネコ", "その他"]
plt.title("好きな動物")
axes[1,2].pie(x5, labels=label)
axes[1,2].set_title("円グラフ")

plt.show()
```
